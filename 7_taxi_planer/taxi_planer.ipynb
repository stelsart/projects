## Обучение

```python
def rmse(predictions, targets):
    return np.sqrt(np.mean((predictions - targets) ** 2))
```

```python
scorer = make_scorer(rmse, greater_is_better=False)
```

```python
def make_features(dа, max_lag, rolling_mean_size):
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
make_features(df, 10, 2)
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()

features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

```python
model_lr = LinearRegression().fit(features_train, target_train)

tscv = TimeSeriesSplit(n_splits=5)

rmse_lr = cross_val_score(model_lr, features_train, target_train, cv=tscv, scoring=scorer).mean()
print('Лучшее значение remse при линейной регрессии на трейне:', -rmse_lr)
```

```python
model_cb = CatBoostRegressor(random_state = 12345)
parameters = {'depth' : [6,8,10],
              'learning_rate' : [0.01, 0.05, 0.1],
              'iterations'    : [30, 50, 100]
              }

grid = GridSearchCV(estimator=model_cb, param_grid = parameters, cv = tscv, n_jobs=-1, scoring=scorer)
grid.fit(features_train, target_train)

print('Лучшее значение rmse и параметров CatBoostRegressor на тесте:',
      -grid.best_score_, grid.best_params_)
```

```python
model_rf = RandomForestRegressor(random_state = 12345)
parameters_rf = {'max_depth' : [6,8,10],
              'n_estimators' : [100, 500, 1000]
              }

grid_rf = GridSearchCV(estimator=model_rf, param_grid = parameters_rf, cv = tscv, n_jobs=-1, scoring=scorer)
grid_rf.fit(features_train, target_train)

print('Лучшее значение rmse и параметров CatBoostRegressor на тесте:',
      -grid_rf.best_score_, grid_rf.best_params_)
```

Былы проверены на тесте с кросс валидацией 3 модели регрессии - линейная, случайный лес и catboost. Лучше значение rmse = 26.02 на тесте получилось у catboost c парметрами depth': 8, 'iterations': 100, 'learning_rate': 0.1
